```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from utils import clean_dataframe
```

```python
df = pd.read_csv("bouts_out_new.csv")
```

```python
df.shape
```

```python
df = clean_dataframe(df)
```

```python
df.shape
```

```python
df["diff_kos_a_b"] = df.kos_a - df.kos_b
```

```python
df.head()
```

```python
ax = df[df.gano==1].plot.scatter(x='kos_a', y='kos_b', color='DarkBlue', label='Gano',figsize=(16,16));
df[df.gano==0].plot.scatter(x='kos_a', y='kos_b', color='DarkGreen', label='No gano', ax=ax,  figsize=(16,16));
```

```python
f,ax=plt.subplots(1,2,figsize=(16,6))
ax[0].set_title('# de victorias cuando A>B')
df[df.diff_kos_a_b>0].gano.value_counts().plot.bar(ax=ax[0])
ax[1].set_title('% de victorias cuando A>B')
(df[df.diff_kos_a_b > 0].gano.value_counts().sort_index() / df.gano.value_counts().sort_index()).plot.bar(ax=ax[1])
```
